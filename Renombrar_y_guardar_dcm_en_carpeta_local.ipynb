```python
path = "../../ProstateX/Prostatex_test/PROSTATEx/"
```

```python
import os

os.makedirs("resultados", exist_ok=True)

carpetas = os.listdir(path)
for carpeta in carpetas:
    id_paciente = carpeta[-4:]
    aux = os.path.join(path,carpetas[0])
    aux = os.path.join(aux,os.listdir(aux)[0])
    aux2 = os.listdir(aux)
    aux = os.path.join(aux, [b for b in aux2 if "t2tsetra" in b][0])
    ficheros_paciente = os.listdir(aux)
    print("Ficheros del paciente {}:".format(id_paciente))
    print(ficheros_paciente)
    for fichero in ficheros_paciente:
        fichero_origen = "'{}'".format(os.path.join(aux,fichero))
        fichero_destino = "./resultados/" + id_paciente + "_" + fichero
        fichero_destino = "'{}'".format(fichero_destino)
        comando = "cp {} {}".format(fichero_origen, fichero_destino)
        os.system(comando)
        
```
